```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_sales_2020_2021_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], join = 'inner', axis ='rows')
athletic_sales_2020_2021_df.reset_index(inplace=True)
athletic_sales_2020_2021_df
```

```python
# Check if any values are null.
athletic_sales_2020_2021_df.isnull()
```

```python
athletic_sales_2020_2021_df.isnull().sum()
```

```python
# Check the data type of each column
athletic_sales_2020_2021_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
athletic_sales_2020_2021_df['invoice_date'] = pd.to_datetime(athletic_sales_2020_2021_df['invoice_date'] , format = '$m/$d/$Y')
```

```python
# Confirm that the "invoice_date" data type has been changed.
athletic_sales_2020_2021_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
region_state_city_df = athletic_sales_2020_2021_df.groupby(['region','state','city'])[['units_sold']].sum()
region_state_city_df = region_state_city_df.rename(columns={'units_sold':'Total_Products_Sold'})
regions_state_city_sorted = region_state_city_df.sort_values(by='Total_Products_Sold', ascending = False)
# Show the top 5 results.
regions_state_city_sorted.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_products_pivot_df = pd.pivot_table(athletic_sales_2020_2021_df,
                                   values = 'units_sold',
                                   index = ['region','state','city'],
                                   aggfunc= 'sum')

# Rename the "units_sold" column to "Total_Products_Sold"
total_products_pivot_df = region_state_city_df.rename(columns={'units_sold':'Total_Products_Sold'})
total_products_pivot_sorted = total_products_pivot_df.sort_values(by='Total_Products_Sold', ascending = False)

# Show the top 5 results.
total_products_pivot_sorted.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales_df = athletic_sales_2020_2021_df.groupby(['region','state','city'])[['total_sales']].sum()
total_sales_df = total_sales_df.rename(columns={'total_sales':'Total Sales'})
total_sales_sorted = total_sales_df.sort_values(by='Total Sales', ascending = False)
# Show the top 5 results.
total_sales_sorted.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

total_sales_pivot_df = pd.pivot_table(athletic_sales_2020_2021_df,
                                   values = 'total_sales',
                                   index = ['region','state','city'],
                                   aggfunc= 'sum')
# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_pivot_df = total_sales_pivot_df.rename(columns={'total_sales':'Total Sales'})
total_sales_pivot_sorted = total_sales_pivot_df.sort_values(by='Total Sales', ascending = False)
# Show the top 5 results.
total_sales_pivot_sorted.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales_retailer_df = athletic_sales_2020_2021_df.groupby(['retailer','region','state','city'])[['total_sales']].sum()
total_sales_retailer_df = total_sales_retailer_df.rename(columns={'total_sales':'Total Sales'})
total_sales_retailer_sorted = total_sales_retailer_df.sort_values(by = 'Total Sales', ascending = False)

# Show the top 5 results.
total_sales_retailer_sorted.head()
```

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
total_sales_retailer_pivot_df = pd.pivot_table(athletic_sales_2020_2021_df,
                                   values = 'total_sales',
                                   index = ['retailer','region','state','city'],
                                   aggfunc= 'sum')

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_retailer_pivot_df = total_sales_retailer_pivot_df.rename(columns={'total_sales':'Total Sales'})
total_sales_retailer_pivot_sorted = total_sales_retailer_pivot_df.sort_values(by = 'Total Sales', ascending = False)

# Show the top 5 results.
total_sales_retailer_pivot_sorted.head()
```
